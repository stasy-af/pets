Импорт файла

```python
import pandas as pd

df = pd.read_csv('train.csv', index_col=0)
print(df)
```

Краткий обзор и описание дата сета

```python
print(df.head())
print(df.info())
print(df.describe())
print(df.shape)
```

Из выведенной информации можно сделать вывод, что в датасете присутствуют N/A поля в столбцах с возрастом, номером каюты и портом отправления. Кроме того, большая часть пассажиров было в каютах 3 класса, 75% из них были довольно молодыми (<40), почти все люди на борту не знали друг друга (небольшое количество людей, которые путешествовали в паре или с сиблингами и почти никого, кто был с родителями или детьми)

Второй этап - чистка данных, заполняем возраст средним значением по столбцу, порт отправление заполняем 'S', так как это наиболее часто встречающийся порт, а также удаляем столбцы с данными о билете и каюте

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna('S')
print(df.isna().sum()) # additional check

df = df.drop(['Ticket', 'Cabin'], axis=1)
```

Ответы на вопросы:

Каков общий процент выживших и невыживших?

Как выживаемость зависит от пола (Sex)?

Как выживаемость зависит от класса билета (Pclass)?

Есть ли различия по возрасту: дети/взрослые/пожилые?

Влияет ли цена билета (Fare) на выживаемость?

```python
# question 1

survive_table = pd.Series(index=[0, 1], dtype=float)
total = len(df)
sur_count = df['Survived'].value_counts().sort_index()
survive_table[0] = (sur_count[0] / total) * 100
survive_table[1] = (sur_count[1] / total) * 100
print(survive_table)
```

```python
# question 2

survive_by_gender = df.groupby('Sex')['Survived'].sum().astype(float)
total_sur = df.shape[0]
survive_by_gender.iloc[0] = (survive_by_gender.iloc[0] / total_sur) * 100 # women
survive_by_gender.iloc[1] = (survive_by_gender.iloc[1] / total_sur) * 100 # men
print(survive_by_gender)
```

```python
# question 3

survive_by_class = df.groupby('Pclass')['Survived'].sum().astype(float)
survive_in_class = df.groupby('Pclass')['Survived'].count().astype(float)
survive_in_class.iloc[0] = (survive_by_class.iloc[0] / survive_in_class.iloc[0]) * 100
survive_in_class.iloc[1] = (survive_by_class.iloc[1] / survive_in_class.iloc[1]) * 100
survive_in_class.iloc[2] = (survive_by_class.iloc[2] / survive_in_class.iloc[2]) * 100
total_sur = df.shape[0]
survive_by_class.iloc[0] = (survive_by_class.iloc[0] / total_sur) * 100 # first class
survive_by_class.iloc[1] = (survive_by_class.iloc[1] / total_sur) * 100 # second class
survive_by_class.iloc[2] = (survive_by_class.iloc[2] / total_sur) * 100 # third class
print(survive_by_class) # общий процент выживших в каждом классе
print(survive_in_class) # процент выживших в конкретном классе от общего числа пассажиров класса
```

```python
# question 4

bins = [0, 18, 55, 80]
labels = ['0-18', '18-55', '55-80']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
survival_by_age = df.groupby('AgeGroup', observed=True)['Survived'].sum().astype(float)
survival_in_age_group = df.groupby('AgeGroup', observed=True)['Survived'].count().astype(float)
survival_in_age_group.iloc[0] = (survival_by_age.iloc[0] / survival_in_age_group.iloc[0]) * 100
survival_in_age_group.iloc[1] = (survival_by_age.iloc[1] / survival_in_age_group.iloc[1]) * 100
survival_in_age_group.iloc[2] = (survival_by_age.iloc[2] / survival_in_age_group.iloc[2]) * 100
total_sur = df.shape[0]
survival_by_age.iloc[0] = (survival_by_age.iloc[0] / total_sur) * 100 # age 0-18
survival_by_age.iloc[1] = (survival_by_age.iloc[1] / total_sur) * 100 # age 18-55
survival_by_age.iloc[2] = (survival_by_age.iloc[2] / total_sur) * 100 # age 55-80
print(survival_by_age)
print(survival_in_age_group)
```

```python
# question 5

print(df.groupby('Survived', observed=True)['Fare'].agg(['min', 'median', 'max']))
```

Проанализировав полученные данные, можно сделать следующие выводы:
1) Больше спаслось женщин, чем мужчин
2) Больше спалось людей, которые заплатили больше денег за билет, т.е. из первого и второго классов
3) Больше спаслось детей, чем взрослых и стариков

Исходя из этого, можно выдвинуть гипотезу, что с большей вероятностью спаслась девочка из первого класса, чем пожилой мужчина из третьего класса
